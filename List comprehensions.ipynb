## list comprehension::
## syntax:: [exp for iterator in iterable if condition]

```python
# find 1-n numbers as list using list comprehension::
n=int(input())
res= [x for x in range(1,n+1) ]
print(res)
```

```python
#find the first n even numbers:
n=int(input())
res = [2*x for x in range(n)]
print(res)
```

```python
#find the first n even numbers:
n=int(input())
res = [x for x in range(0,2*n,1) if x %2 == 0]
print(res)
```

```python
# return list even or odd, for list an integers:
arr =[13,6,23,46,59,14,96]
res = ["odd"if x&1 else "even" for x in arr] 
print(res)
```

```python
# even-add 1
# odd - add 0.5
arr =[13,6,23,46,59,14,96]
res = [x + 0.5 if x&1 else x+1 for x in arr] 
print(res)
```

```python
# find the one dice combinations:
print([x for x in range(1,7)])
```

```python
# find the two dice combinations:
print([(i,j) for i in range(1,7) for j in range(1,7)])
```

```python
# find the two dice combinations if any dice face value is 4:
print([(i,j) for i in range(1,7) for j in range(1,7) if i == 4 or j == 4])
```
